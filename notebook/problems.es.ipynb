## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
# TODO
import numpy as np

simulaciones = 1000

dado1 = np.random.randint (1, 7, size=simulaciones)
dado2 = np.random.randint (1, 7, size=simulaciones)

suma_dados = dado1 + dado2

probabilidad = (np.sum((suma_dados > 7) | (suma_dados % 2 == 0)) / simulaciones) * 100

print (f"La probabilidad de tirar dos dados 1000 veces y que la suma de ambos sea par o mayor de 7 es de {probabilidad}%")
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Tomar todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
# TODO

simulaciones = 1000

def extraer_bola (n_experimentos):

    contador_3b_2r = 0
    contador_mismo_color = 0

    for i in range (n_experimentos):
        color_bola = []

        for i in range (5):
            color_bola.append(ball_box[np.random.randint(0,59)])

        color_bola = np.array(color_bola)

        bola_blanca = sum(color_bola == "White")
        bola_roja = sum(color_bola == "Red")
        bola_verde = sum(color_bola == "Green")

        if (bola_blanca == 3) and (bola_roja == 2):
            contador_3b_2r += 1
        
        if (bola_blanca == 5) or (bola_roja == 5) or (bola_verde == 5):
            contador_mismo_color += 1
        
    
    return contador_3b_2r, contador_mismo_color
                              

prob_bolas = extraer_bola(simulaciones)

prob_3b_2r = (prob_bolas[0] / 1000) * 100
prob_mismo_color = (prob_bolas[1] / 1000) * 100

print (f"La probabilidad de extraer 3 bolas blancas y 2 bolas rojas es de {prob_3b_2r}%")
print (f"La probabilidad de extraer todas las bolas del mismo color es de {prob_mismo_color}%")




```
